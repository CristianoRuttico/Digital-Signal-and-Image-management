# Generazione dataset segnali audio

```python
#import librerie
import sounddevice as sd
from scipy.io import wavfile as wav
import IPython.display as ipd
import os
```

```python
#creazione cartella
path = os.getcwd()
print ("The current working directory is %s" % path)
```

```python
#creazione cartella che va fatta solo una volta
os.mkdir(r'C:\Users\crutt\dsim\9\cr_sala')
```

```python
#Impostazione parametri di recording
duration = 5 # (seconds)
rec_rate = 44100
```

```python
#Numero di file audio che vuoi ottenere
k=17
```

```python
#registratore e creazione lista in cui inserire le rec: al posto di 'cr' mette vostra iniziale nome e iniziale cognome.
cr_rec=[]
for i in range(k):
    rec = sd.rec(int(duration * rec_rate), samplerate=rec_rate, channels=1, blocking=True)
    cr_rec.append(rec)
```

```python
#Pulizia della cartella: cancella tutte le registrazioni. Nel caso in cui qualcosa sia andato storto durante la registrazione.
#al posto di dir mettete il vostro percorso
dir = r'C:\Users\crutt\dsim\9\cr_sala'
for f in os.listdir(dir):
    os.remove(os.path.join(dir, f))
```

```python
#Inserisce registrazioni nella cartella. Anche qua sostituite a 'cr' vostra iniziale nome e iniziale cognome
num = 0
for audio in cr_rec:
    cr = str(num) +'_'+ 'cr'+'_'
    wav.write(f'cr_sala/{cr}.wav', rate=rec_rate, data=audio)
    num = num + 1    
```
